```python
from statsmodels.tsa.stattools import adfuller 
```

```python
test_result = adfuller(df['Close']) 
```

```python
# ho: The data is not stationary.
# ha: The data is stationary.
```

```python
# Creating a function to check my assumption
def adfullet_test(Close):
    result = adfuller(Close)
    labels = ['ADF Test', 'P-value', 'Lags Used', 'Number Of Observation']
    for value, label in zip(result, labels):
        print(label + ': ' + str(value))
    if result[1] <= 0.05:
        print("Null Hypothesis is rejected. The data is Stationary")
    else:
        print("Null Hypothesis is accepted. The data is not Stationary")
```

```python
adfullet_test(df['Close'])
```

```python
# The assumption is rejected by very good difference. (0.99 !<= 0.05)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings('ignore')
```

```python
data2 = df[['Close']]
data2
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
data2['Close'] = scale.fit_transform(data2[['Close']])
```

```python
model = SARIMAX(data2, order=(1,1,1), seasonal_order=(1,1,1,12))
model
```

```python
model_fit = model.fit(disp=False) 
print(model_fit.summary()) 
```

```python
forecast_steps = 30
forecast = model_fit.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data2.index[-1], periods=forecast_steps + 1, freq='D')[1:]
forecast_values = scale.inverse_transform(forecast.predicted_mean.values.reshape(-1, 1))
```

```python
forecast_df = pd.DataFrame(data=forecast_values, index=forecast_index, columns=['Forecast'])
forecast_df
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(forecast_df, s=100)
plt.xlabel('Date')
plt.ylabel('Forecast')
plt.title('Forecast Data')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(forecast_df, color='orange', linestyle='--', marker='*')
plt.xlabel('Date')
plt.ylabel('Forecast')
plt.title('Forecast Data')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data2.index, scale.inverse_transform(data2), label='Actual')
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecast')
plt.title('Reliance Stock Price Prediction for Next 30 Days using SARIMAX')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
